```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración de estilo para gráficos más estéticos
sns.set_style("whitegrid")
```

```python
import numpy as np

# Generar datos simulados
datos = np.random.normal(loc=50, scale=15, size=200)  # Media 50, Desviación 15

# Crear histograma con Matplotlib
plt.figure(figsize=(8,5))
plt.hist(datos, bins=20, color="skyblue", edgecolor="black", alpha=0.7)
plt.title("Distribución de Datos")
plt.xlabel("Valores")
plt.ylabel("Frecuencia")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(datos, bins=20, kde=True, color="purple")
plt.title("Histograma con KDE")
plt.xlabel("Valores")
plt.ylabel("Frecuencia")
plt.show()
```

```python
# Crear datos simulados
import pandas as pd

df = pd.DataFrame({
    "Categoria": np.random.choice(["A", "B", "C"], size=100),
    "Valores": np.random.normal(50, 15, 100)
})

# Crear diagrama de caja
plt.figure(figsize=(8,5))
sns.boxplot(x="Categoria", y="Valores", data=df, palette="Set2")
plt.title("Diagrama de Caja por Categoría")
plt.xlabel("Categoría")
plt.ylabel("Valores")
plt.show()
```

```python
# Datos simulados
np.random.seed(42)
precio = np.random.normal(200000, 50000, 100)  # Precio en dólares
tamano = precio / 1000 + np.random.normal(0, 20, 100)  # Tamaño en metros cuadrados

# Crear DataFrame
df = pd.DataFrame({"Precio": precio, "Tamaño": tamano})

# Crear scatter plot con Seaborn
plt.figure(figsize=(8,5))
sns.scatterplot(x="Tamaño", y="Precio", data=df, color="blue")
plt.title("Relación entre Tamaño y Precio de Casas")
plt.xlabel("Tamaño (m²)")
plt.ylabel("Precio ($)")
plt.show()
```

```python
import pandas as pd
import numpy as np

# Crear datos para el DataFrame
nombres = ['Empleado ' + str(i) for i in range(1, 101)]
industrias = np.random.choice(['Tecnología', 'Finanzas', 'Salud', 'Educación', 'Retail'], size=100)
sueldos = np.random.randint(30000, 150000, size=100)

# Crear el DataFrame
df_salarios = pd.DataFrame({
    'nombre_empleado': nombres,
    'industria': industrias,
    'sueldo': sueldos
})

# Mostrar las primeras 5 filas del DataFrame
print(df_salarios.head())
```

```python
# Histogramas de salarios
plt.figure(figsize=(8,5))
sns.histplot(df_salarios["sueldo"], bins=30, kde=True, color="green")
plt.title("Distribución de Salarios")
plt.xlabel("Salario")
plt.ylabel("Frecuencia")
plt.show()

# Boxplot por industria
plt.figure(figsize=(10,5))
sns.boxplot(x="industria", y="sueldo", data=df_salarios, palette="coolwarm")
plt.xticks(rotation=45)
plt.title("Salarios por Industria")
plt.xlabel("Industria")
plt.ylabel("Salario")
plt.show()
```
