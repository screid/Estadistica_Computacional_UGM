# 5. Tratamiento de datos nulos

```python
df.isnull().sum()
```

```python
# Eliminemos la columna Publisher
df_sin_publisher = df.dropna()
```

```python
df.shape
```

```python
df_sin_publisher.shape
```

```python
# Rellenar columnas con un valor (generalmente la media)
media = df["Year"].mean()
minimo = df["Year"].min()
df_sin_publisher["Year"].fillna(media, inplace=True)
```

```python
df_sin_publisher
```

```python
# Comprobemos que no hay datos faltantes
df_sin_publisher.isnull().sum()
```

# 6. Transformacion de datos

```python
# Crear una columna nueva que sea la diferencia entre la columna Global_Sales y la columna NA_Sales
df_sin_publisher["Diferencia_Global_NA"] = df_sin_publisher["Global_Sales"] - df_sin_publisher["NA_Sales"]
```

```python
df_sin_publisher.head()
```

# 7. Confirmar duplicados

```python
df_sin_publisher.duplicated().sum()
```

```python
# Si hubiesen datos duplicados, se eliminan mediante esta linea de codigo
df_sin_duplicados = df_sin_publisher.drop_duplicates()
```

```python
df_sin_duplicados.shape
```

# 8. Agregar datos

```python
# Vamos a seleccionar solamente las columnas de tipo numerico
df_numerico = df_sin_duplicados.select_dtypes(include=[np.number])

# Adicionamos la columna Platform a df_numerico
df_numerico["Platform"] = df_sin_duplicados["Platform"]
```

```python
# Voy a agrupar la informacion por la columna "Platform"
df_numerico.groupby("Platform").mean()
```
